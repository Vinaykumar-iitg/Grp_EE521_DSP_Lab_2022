```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax=plt.subplots(4,3,figsize=(20,30))        
n=np.arange(-10,10,1)
x=np.zeros(len(n))
#UNIT STEP SIGNAL(X1):
def unit(a,b,c):
    for i in range(0,len(n)):
        if(n[i]/b)<c:
            x[i]=0
        elif (n[i]/b)>=c and (n[i]/b)<=10+c:
            x[i]=a
    return x        
ax[0][0].stem(n,unit(1,1,0),'r')
ax[0][0].set_title("X1 signal")
ax[0][0].grid() 
#RAMP STEP SIGNAL(X1):
def ramp(a,b,c):
    for i in range(0,len(n)):
        if (n[i]/b)<=c:
            x[i]=0
        elif n[i]/b>c :
            x[i]=a*((n[i]/b)-c)
    return x
ax[0][1].stem(n,ramp(1,1,0),'y')
ax[0][1].set_title("X2 signal")
ax[0][1].grid()

#signal x3:
def par(a,b,c):
    for i in range(0,len(n)):
        if n[i]/b<=c:
            x[i]=0
        elif n[i]/b>c :
            x[i]= a*(((n[i]/b)-(c/b))**2)/2
    return x    
ax[0][2].stem(n,par(1,1,0),'g')
ax[0][2].set_title("X3 signal")
ax[0][2].grid()
tau=5
def rect(a,b,c):
    for i in range(0,len(n)):
        if n[i]>=(-tau/2)+c and np.abs(n[i])<=(tau/2)+c:
            x[i]=a
        else:
            x[i]=0
    return x    
ax[1][0].stem(n,rect(1,1,0),'orange')
ax[1][0].set_title("X4 signal")
ax[1][0].grid() 
#x5 signal:
x4=rect(1,1,0)
x5=np.convolve(x4,x4)
ax[1][1].stem(np.arange(-20,19,1),x5,'r')
ax[1][1].set_title("X5 signal")
ax[1][1].grid() 
#X6 signal:
n1=np.linspace(-10,10,100)
x6= np.sin(3.14*n1)/(3.14*n1)
ax[1][2].stem(n1,x6,'purple')
ax[1][2].set_title("X6 signal")
ax[1][2].grid() 

 
d1,d2,d3=3,10,15
N=np.arange(-5,20,1)
x7=np.zeros(len(N))
for i in range(0,len(N)):
    if N[i]<d1:
        x7[i]=0
    elif N[i]>=d1 and N[i]<d2:
        x7[i]=(N[i]-d1)/(d2-d1)
    elif N[i]>=d2 and N[i]<d3:
        x7[i]=(d3-N[i])/(d3-d2) 
    elif  N[i]>=d3 :
        x7[i]=0
ax[2][0].stem(N,x7,'green')
ax[2][0].set_title("X7 signal")
ax[2][0].grid()   

    
    
ax[2][1].stem(n,unit(4,1,2),'r')
ax[2][1].set_title("X1 signal")
ax[2][1].grid()     
ax[2][2].stem(n,ramp(4,1,2),'y')
ax[2][2].set_title("X2 signal")
ax[2][2].grid()
ax[3][0].stem(n,par(4,1,2),'g')
ax[3][0].set_title("X3 signal")
ax[3][0].grid()
ax[3][1].stem(n,rect(4,1,2),'orange')
ax[3][1].set_title("X4 signal")
ax[3][1].grid() 
x4=rect(4,1,2)
x5=np.convolve(x4,x4)
ax[3][2].stem(np.arange(-20,19,1),x5,'r')
ax[3][2].set_title("X5 signal")
ax[3][2].grid() 
```

## Triangular Wave

```python
d1,d2,d3=0,3,6
n=np.arange(-15,25,1)
x=np.zeros(len(n))
fig, ax=plt.subplots(1,2,figsize=(12,7))        
def tri(a,c):
    for i in range(0,len(n)):
        if n[i]<d1+c:
            x[i]=0
        elif n[i]>=d1+c and n[i]<d2+c:
            x[i]=a*(n[i]-c-d1)/(d2-d1)
        elif n[i]>=d2+c and n[i]<d3+c:
            x[i]=a*(d3-(n[i]-c))/(d3-d2) 
        elif  n[i]>=d3+c :
            x[i]=0
    return x
T=np.array(tri(-1,-18))+np.array(tri(1,-12))+np.array(tri(-1,-6))+np.array(tri(1,0))+np.array(tri(-1,6))+np.array(tri(1,12))+np.array(tri(-1,18))
ax[0].plot(n,T,'green')
ax[0].set_title("Alternating triangular wave ")
ax[0].grid()   

#GENERATING IMPULSE TRAIN :
xmn=np.zeros(len(n))
for i in range(0,len(n)):
    if n[i]<=100 and n[i]>=-100:
        xmn[i]=1       
    else:
        xmn[i]=0
#sampling continuous triangular signal:
triangle=T*xmn
ax[1].stem(n,triangle,'orange')
ax[1].set_title("Sampled triangular signal using impulse train")
ax[1].grid()
```
