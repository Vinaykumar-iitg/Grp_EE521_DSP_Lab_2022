```python
def step(x):
    
    if x>=0:
        return 1
    else:
        return 0
def g_step(x):
    
    fun_val=[]
    for i in x:
        fun_val.append(step(i))
        
    return fun_val
```

## Exponential signals

```python
def fun(A,B):
    L=[]
    t=np.arange(-2,3,0.01)
    
    for i in range(len(t)):
        L.append(A*np.exp(-B*t[i]))
    
    
    return L
```

#### Let x[n] be the input to an linear-time invariant (LTI) system and h[n] be the impulse response. The output of a discrete LTI system in terms of the impulse response is characterized by the equation as y[n] = x[n] ∗ h[n] . Write a function for performing convolution of two sequences.

```python
def fun(ans,l,k):
    a=l
    b=k
    s=0
    while a<len(ans) and b>=0:
        s=s+ans[a][b]
        a=a+1
        b=b-1
    return s
```

```python
def convolve(h,x):
   
    L=[]
    M=[]
    ans=[]
    for j in range(len(h)):
        for i in range(len(x)):
            L.append(h[j]*x[i])
        
        M.append(L)
        L=[]
#     print(M)
    
    def fun(ans,l,k):
        a=l
        b=k
        s=0
        while a<len(ans) and b>=0:
            s=s+ans[a][b]
            a=a+1
            b=b-1
        return s
    
    for i in range(len(M[0])):
        ans.append(fun(M,0,i))
    for j in range(1,len(M)):
        ans.append(fun(M,j,len(M[0])-1))
    
    return ans
```
