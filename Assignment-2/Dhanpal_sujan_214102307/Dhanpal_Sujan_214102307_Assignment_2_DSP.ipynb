```python
def binomial(n,x,p,q):
    s1 = np.math.factorial(n)/(np.math.factorial(x)*np.math.factorial(n-x));
    s2 = p**x;
    s3 = q**(n-x);
    s4 = s1*s2*s3;
    return s4;

without_hamming_code=binomial(4,0,0.1,0.9)
with_hamming_code= binomial(7,0,0.1,0.9)+binomial(7,1,0.1,0.9)

Data = {'using error-correcting':with_hamming_code,'without using error-correcting':without_hamming_code}
names = list(Data.keys())
values = list(Data.values())
plt.bar(names,values)
```

```python
def corr_coef(a,b):
    a1=[]
    b1=[]
    num=0
    den1=0
    den2=0
    mean_a=(np.sum(a))/(len(a))
    mean_b=(np.sum(b))/(len(b))
    
    for i in a:
        a1.append(i-mean_a)
    
   
    for i in b:
        b1.append(i-mean_b)
    
   
    for i in range(len(a1)):
        
        num=num+a1[i]*b1[i]
        den1=den1+((a1[i])**2)
        den2=den2+((b1[i])**2)
    
    return num/np.sqrt(den1*den2)
```

```python
def cov(a,b):
    
    a1=[]
    b1=[]
    num=0
    mean_a=(np.sum(a))/(len(a))
    mean_b=(np.sum(b))/(len(b))
    
    for i in a:
        a1.append(i-mean_a)
    
   
    for i in b:
        b1.append(i-mean_b)
    
    for i in range(len(a1)):
        num=num+a1[i]*b1[i]
        
    return num/(len(a1)-1)
```
