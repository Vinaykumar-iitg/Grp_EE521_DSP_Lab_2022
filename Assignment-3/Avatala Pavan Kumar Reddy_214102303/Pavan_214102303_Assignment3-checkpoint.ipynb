## 1.

#### Find the Fourier series associated with the function
 $$ f(x)  =\begin{cases} 
    0 & \text{ -pi $\leq$ x <0}\\
    \pi-x & \text{0 $\leq$ $t$ $\leq\pi$ }\\
    \end{cases} $$
####  Write a function to determine the coefficients $a_{0}$, $a_{n}$, and $b_{n}$. Plot themagnitude of the coefficients with |$a_{n}$| in blue colour and |$b_{n}$| in redcolour.

```python
T=2*np.pi #time perid of the function
w=2*np.pi/T 
def a(n): 
    f = lambda x:(np.pi-x)*np.cos(n*w*x) if (0<=x<=np.pi) else 0 #formula for an
    k= integrate.quad(f,-np.pi, np.pi)
    return round(k[0]*2/T,2)
def b(n): 
    f = lambda x:(np.pi-x)*np.sin(n*w*x) if (0<=x<=np.pi) else 0 #formula for bn
    k= integrate.quad(f,-np.pi, np.pi)
    return round(k[0]*2/T,2)


print('The average value a0 is :',a(0)/2)
```

```python
plt.figure(1,figsize=(14,7))
n=np.arange(1,15,1)

A=[a(i) for i in n]

plt.subplot(2,1,1)
plt.stem(n,A,'b')
plt.xlabel('n')
plt.ylabel("magnitude")
plt.title('$|a_{n}|$',size=15)

B=np.abs([b(i) for i in n])
plt.subplot(2,1,2)
plt.stem(n,B,'r')
plt.xlabel('n')
plt.ylabel("magnitude")
plt.title('$|b_{n}|$',size=15)

plt.show()
```

##### (b) Write a function for discrete Fourier transform (DFT) as DFT(x) which takes in an input argument x, which is a 1-dimensional real valued signal. The function has to calculate the DFT of the signal and return the DFT values. Apply this function to the signal generated in (a) and plot the result. Verify that the output of the DFT is symmetric.
##### (c) Filter out the DFT results to obtain only the frequency components of (a) and plot the resultant. That is, filter the frequency components which is greater than fs/2.

```python
# sampling rate
fs=100
# sampling interval
ts=1.0/fs
t=np.arange(0,1,ts)

#frequency=2
x1=3*np.sin(2*np.pi*2*t)

#frequency=5
x2=2*np.sin(2*np.pi*5*t)

#frequency=8   
x3=np.sin(2*np.pi*8*t)

x=x1+x2+x3

plt.figure(figsize = (14, 7))
plt.stem(t, x)
plt.ylabel('Amplitude')


plt.show()
```

```python
def DFT(x):
    N=len(x)
    n=np.arange(N)
    k=n.reshape((N, 1))
    w=np.exp(-2j*np.pi*k*n/N)
    X=np.dot(w,x)    
    return X
```

```python
X=DFT(x)

# calculate the frequency
N=len(X)
n=np.arange(N)
T=N/fs
freq=n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X))
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```

```python
n_oneside = N//2
# get the one side frequency
f_oneside = freq[:n_oneside]

# normalize the amplitude
X_oneside =X[:n_oneside]/n_oneside

plt.figure(figsize = (12, 6))
plt.stem(f_oneside, abs(X_oneside))
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')

plt.show()
```
